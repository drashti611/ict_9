```python
class Car:
    def __init__(self,brand,model,year,price,fuel_type,color):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price
        self.fuel_type = fuel_type
        self.color = color
    
    def displayInfo(self):
        print(f"Brand      : {self.brand}")
        print(f"Model      : {self.model}")
        print(f"Year       : {self.year}")
        print(f"Price      : Rs.{self.price}")
        print(f"Fuel Type  : {self.fuel_type}")
        print(f"Color      : {self.color}")
        print("---------------------------")
        
class ShowRoom:
    def __init__(self):
        self.inventory = []
        
    def viewAvailableCars(self):
        if not self.inventory:
            print("No cars available in the showroom.")
            return
        print("\nAvailable Cars: ")
        for idx,car in enumerate(self.inventory,start=1):
            print(f"{idx}. {car.brand} {car.model} ({car.year}) - Rs.{car.price}")
        print()
        
    def displayCarDetails(self,index):
        if 0 <= index < len(self.inventory):
            print("\nCar Details: ")
            self.inventory[index].displayInfo()
        else:
            print("Invalid Car Index.\n")
    
    def sellCar(self,index):
        if 0 <= index < len(self.inventory):
            soldCar = self.inventory.pop(index)
            print(f"\nCar Sold: {soldCar.brand} {soldCar.model} ({soldCar.year})")
        else:
            print("Invalid car index. Cannot sell.\n")
    
    def buy_car(self, car):
        self.inventory.append(car)
        print(f"\nCar Purchased and Added to Showroom: {car.brand} {car.model} ({car.year})\n")

def main():
    
    showroom = ShowRoom()
    
    showroom.buy_car(Car("Toyota", "Camry", "2020", "2500000", "Petrol", "White"))
    showroom.buy_car(Car("Hyundai", "Creta", "2022", "1800000", "Diesel", "Black"))
    showroom.buy_car(Car("Tata", "Nexon", "2023", "1400000", "Electric", "Blue"))
    
    while True:
        print("\n====== Car Showroom Menu ======")
        print("1. View Available Cars")
        print("2. Display Car Details")
        print("3. Sell a Car")
        print("4. Buy a Car")
        print("5. Exit")
        choice = input("Enter Your Choice(1-5): ")
        
        if choice == '1':
            showroom.viewAvailableCars()
            
        elif choice == '2':
          showroom.viewAvailableCars()
            
          index = int(input("Enter car number to view details: ")) - 1
          showroom.displayCarDetails(index)
           
  
        elif choice == '3':
            showroom.viewAvailableCars()
            try:
                index = int(input("Enter car number to sell: ")) - 1
                showroom.sellCar(index)
            except ValueError:
                print("Please enter a valid number.")

        elif choice == '4':
            print("\nEnter car details to buy:")
            brand = input("Brand: ")
            model = input("Model: ")
            try:
                year = int(input("Year: "))
                price = int(input("Price (INR): "))
            except ValueError:
                print("Year and price must be numbers.")
                continue
            fuel_type = input("Fuel Type: ")
            color = input("Color: ")

            new_car = Car(brand, model, year, price, fuel_type, color)
            showroom.buy_car(new_car)
        
        elif choice == '5':
            print("Exiting the program. Thank you!")
            break

        else:
            print("Invalid choice. Please try again.")
            
if __name__ == "__main__":
    main()
```
